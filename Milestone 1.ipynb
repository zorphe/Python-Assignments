```python
from IPython.display import clear_output
import random
#clear_output()

test_board = ['#','X','O','X','O','X','O','X','O','X']
test_board_two = ['#', 'X', 'O', 'X', 'O', 'X', 'O', 'X', 'O', '']

def display_board(board):
    '''
        INPUT: board []
        OUPUT: VISUAL ASCII BOARD
    '''
    def print_value_line(a,b,c):
        return print(f' {a} | {b} | {c} ')
    def print_empty_line():
        return print('   |   |   ')
    def print_sep_line():
        return print('---|---|---')
    
    clear_output()
    
    print_empty_line()
    print_value_line(board[7],board[8],board[9])
    print_empty_line()
    print_sep_line()
    print_empty_line()
    print_value_line(board[4],board[5],board[6])
    print_empty_line()
    print_sep_line()
    print_empty_line()
    print_value_line(board[1],board[2],board[3])
    print_empty_line()

def get_first_player_marker():
    marker = False
    while not marker:
        marker = input("Please pick a marker 'X' or 'O'").upper()
        if marker == 'X' or marker == 'O':
            return marker
        else:
            marker = False
            
def place_marker(board, marker, position):
    board[position] = marker

def check_board(board, mark):
    def check_line(board, a):
        return board[a[0]] != ' ' and board[a[0]] == board[a[1]] and board[a[0]] == board[a[2]]
    
    checks = [
        [1,2,3], [4,5,6], [7,8,9],
        [1,4,7], [2,5,8], [3,6,9],
        [1,5,9], [3,5,7]
    ]
    
    for check in checks:
        if check_line(board, check):
            return mark
    return False

def found_unoccupied_space(board, position):
    return board[position] == ' '

def full_board_check(board):
    return " " not in board[1:]
    #for i in range(1,10):
    #    if (found_unoccupied_space(board, i)):
    #        return False
    #return True

def choose_first():
    if random.randint(0, 1) == 1:
        return 'X'
    else:
        return 'O'

def player_one_marker():
    choice = ''
    while not (choice == 'X' or choice == 'O'):
        choice = input('Player 1: choose "X" or "O"').upper()
        
    if choice == 'X':
        alternative = 'O'
    else:
        alternative = 'O'
    return (choice, alternative)
    
    
def player_choice(board):
    position = False
    while not position:
        try:
            position = int(input("\nPlease pick a location"))
        except ValueError:
            print('Sorry, please choose a number between 1 - 9.')
            continue
            
        if found_unoccupied_space(board, position):
            return position
        else:
            position = False
            
def play_another_game():
    response = input("Would you like to play another game? (Y/N)")
    return response.lower() == 'y'
```

```python
print('Welcome to Tic Tac Toe!')

player_one_marker, player_two_marker = player_one_marker()

while True:
    # Set the game up here
    firstplayer = choose_first()
    if firstplayer == 'X':
        secondplayer = 'O'
    else:
        secondplayer = 'X'
    print(f'{firstplayer} will go first.')
    
    turn = firstplayer 
    board = [' ']*10
    winner = False
    
    while not winner:
        #Player 1 Turn   
        if turn == firstplayer:
            display_board(board)
            p1 = player_choice(board)
            place_marker(board, firstplayer, p1)
            display_board(board)

            winner = check_board(board, firstplayer)
            if winner or full_board_check(board):
                break
            turn = secondplayer
        
        # Player2's turn.
        else:
            display_board(board)  
            p2 = player_choice(board)
            place_marker(board, secondplayer, p2)

            winner = check_board(board, secondplayer)
            if full_board_check(board):
                break
                
            turn = firstplayer

    if winner:
        display_board(board)
        print(f'{winner} won the game!')
    else:
        print('Tied game. Better luck next time!')
        
    if not play_another_game():
        clear_output()
        print('Thanks for playing!')
        break
```
